```python
from textgenrnn import textgenrnn
import feather
import time
```

```python
df = feather.read_dataframe('../data/beer-info.feather')
df.head()
```

```python
textgen = textgenrnn()
textgen.generate()
```

```python
#textgen.train_from_file('hacker-news-2000.txt', num_epochs=1)
start_time = time.time()

textgen.train_on_texts(df.name)

print('Training time:', time.time() - start_time, 'seconds')
```

```python
textgen.generate()
```

```python
new_beer_names = set()
for i in range(20):
    new_beer_names = new_beer_names.union(set(textgen.generate(n=500, return_as_list=True, temperature=0.3)))
    print(len(new_beer_names))
```

```python
new_beer_names = new_beer_names - set(df.name)
```

```python
print(new_beer_names)
```

```python
start_time = time.time()

brewer_gen = textgenrnn()
brewer_gen.train_on_texts(df.brewery)

print('Training time:', time.time() - start_time, 'seconds')

brewery_names = set()
for i in range(20):
    brewery_names = brewery_names.union(set(brewer_gen.generate(n=500, return_as_list=True, temperature=0.3)))
    print(len(brewery_names))
    
brewery_names = brewery_names - set(df.brewery)
print(brewery_names)
```
