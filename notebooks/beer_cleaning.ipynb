```python
import numpy as np
import pandas as pd
```

```python
t = pd.read_json('../data/reviews.json', orient='records')
t.rename(index=str, columns={'rating':'user_rating'}, inplace=True)
t.user_id = t.user_id.apply(lambda v: v.split('/')[-1])
#t.set_index('id', inplace=True)
t.head()
```

```python
def split_ibu(val):
    try:
        newVal = int(val.split(' ')[0])
    except ValueError:
        newVal = np.nan
    return newVal
```

```python
f = pd.read_json('../data/beer-info.json', orient='records')
f.rename(index=str, columns={'rating':'community_rating', 'raters':'num_ratings', 'style':'beer_style'}, inplace=True)
f.community_rating = f.community_rating.apply(lambda v: float(v[1:-1]))
#f.abv = f.abv.apply(lambda v: float(v.split("%")[0])/100 )
f.name = f.name.apply(lambda v: v.split('\n')[0])
f.date = pd.to_datetime(f.date.apply(lambda v: v.split(' ')[1]))
f.num_ratings = f.num_ratings.apply(lambda v: v.split(' ')[0])
f.num_ratings = f.num_ratings.apply(lambda v: int(v.replace(',', '')))
f[['beer_style', 'sub_style']] = f.beer_style.str.split(' - ', expand=True)
f.ibu = f.ibu.apply(split_ibu)
#f.set_index('id', inplace=True)
f.head()
```

```python
a = pd.merge(t, f, how='left', left_on='beer_id', right_on='id')
a
```
